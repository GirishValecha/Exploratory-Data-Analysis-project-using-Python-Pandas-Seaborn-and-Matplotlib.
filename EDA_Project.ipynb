```python
#4. is the distribution of sugar prices in the US significantly different from a normal distribution?
sns.histplot(df["sugar_us"], kde=True)

# Bell-shaped curve → Normal

# Skewed / irregular shape → Not normal

# This is the most common first step in EDA.
```

```python
# shapiro wilk test - test the normality
from scipy.stats import shapiro
# null hypothesis - Data is normally distributed
# alternate hypothesis - Data is not normally distributed

stat, pvalue = shapiro(df["sugar_us"])

if pvalue < 0.05:
    print("Data is not normally distributed")
else:
    print("Data is normally distributed")
```

```python
#9. Is there any significant difference in the distribution of sugar prices between europe(eu) and the united states(us)?
sns.histplot(df['sugar_eu'], color='orange', kde=True, label='EU', bins=15)
sns.histplot(df['sugar_us'], color='blue', kde=True, label='US', bins=15)

plt.title("Distribution of Sugar Prices: EU vs US")
plt.xlabel("Sugar Price")
plt.ylabel("Frequency")

plt.legend()
plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.kdeplot(df['sugar_eu'], fill=True, label='EU')
sns.kdeplot(df['sugar_us'], fill=True, label='US')

plt.title("Sugar Price Distribution: EU vs US")
plt.xlabel("Sugar Price")
plt.legend()

plt.show()
```

```python
#10. how has brent oil prices vary on a quaterly basis since the last five years?
last_5_years = df[df['date'] >= df['date'].max() - pd.DateOffset(years=5)]
last_5_years = last_5_years.set_index('date')

quarterly_brent = last_5_years['oil_brent'].resample('Q').mean()
```

```python
plt.figure(figsize=(7,5))

plt.plot(quarterly_brent, marker='o')

plt.title("Quarterly Brent Oil Price (Last 5 Years)")
plt.xlabel("Quarter")
plt.ylabel("Average Brent Oil Price")

plt.show()
```
